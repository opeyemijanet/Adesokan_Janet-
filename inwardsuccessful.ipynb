```python
import zipfile
import os

# Path to the folder containing the zip files
zip_folder = r"/content/C:\path\to\your\extracted_files/OCTOBER 2025"  # folder where your BOWEN MFB_*.zip files are

# Destination folder for the unzipped files
output_folder = r"C:\path\to\unzipped_files"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Loop through all files in the zip folder
for file_name in os.listdir(zip_folder):
    if file_name.endswith(".zip"):
        zip_path = os.path.join(zip_folder, file_name)

        # Create a subfolder for each zip (optional, keeps things organized)
        subfolder_name = os.path.splitext(file_name)[0]
        subfolder_path = os.path.join(output_folder, subfolder_name)
        os.makedirs(subfolder_path, exist_ok=True)

        # Extract the zip
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(subfolder_path)

        print(f"Extracted: {file_name} → {subfolder_path}")

print("All zip files extracted!")
```

```python
import os
import shutil

# Root folder where all extracted Bowen folders are
unzipped_root = r"/content/C:\path\to\unzipped_files"

# Destination folder
destination = r"/content/C:\path\to\NIP"
os.makedirs(destination, exist_ok=True)

# Loop through each Bowen folder
for folder_name in os.listdir(unzipped_root):
    folder_path = os.path.join(unzipped_root, folder_name)
    if os.path.isdir(folder_path):
        # Walk recursively through this folder
        found = False
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                if file == r"NIP\NIP_617_inwards successful.csv":
                    src_file = os.path.join(root, file)
                    # Rename the file to include parent Bowen folder name
                    dest_file = os.path.join(destination, folder_name + "NIP_617_inwards successful.csv")
                    shutil.copy2(src_file, dest_file)
                    print(f"Copied: {src_file} → {dest_file}")
                    found = True
                    break  # stop searching in this folder once found
            if found:
                break
        if not found:
            print(f"File not found in {folder_name}")

print("All files processed!")
```
